```python
import PyPDF2 as pdf
#pip version: googletrans==4.0.0-rc1 
#pip install PyPDF2

from googletrans import Translator
import matplotlib.pyplot as plt
import numpy as np


def preprocess_text_per_new_line(filepath, page_num):
    #Array declaration###############################
    translation_src = []
    translation_dest = []
    origin_sentences = []
    translation_text = []
    translation_confi = []


    #open pdf file##########################################################
    pdf_file = open(filepath, "rb")
    pdf_reader = pdf.PdfReader(pdf_file)
    #print(f" We got {len(pdf_reader.pages)} pages found in this pdf file!!!") 
    
    #extract text from pdf##################################################    
    selected_page = pdf_reader.pages[page_num]
    new_sentence = selected_page.extract_text()


    
    # split all new line and regiter it #######################################
    for line in new_sentence.splitlines():

        #Translate extracted text######################################
        translator = Translator(service_urls=['translate.google.com'])
        translation = translator.translate(str(line), dest='en') 
    
        #Input result to array###########################
        translation_src.append(translation.src)        
        translation_dest.append(translation.dest)
        origin_sentences.append(line)
        translation_text.append(translation.text)
        translation_confi.append(translator.detect(line).confidence)
        
    #list of data####################################################
    translation_data = {
                        "src_language": translation_src,
                        "dest_language": translation_dest, 
                        "origin_sentences": origin_sentences,
                        "translated_text": translation_text,
                        "translation_confidence": translation_confi
                        }
    return translation_data
```
